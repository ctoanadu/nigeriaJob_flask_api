```python
def get_info(card_one):
    inside=card_one.h5.a
    #url
    href=inside.get('href')
    link='https://www.nigeriajob.com' 
    url_link=link+href
    #job
    job_title=inside.text
    #company
    company_name=card_one.p.b.text
    #job_description
    job_description=card_one.div.text
    #location
    location=card_one.find_all('p')[-1].text
    location=location.split()[3:] 
    location=list(filter(lambda x:x!='-',location))
    
    #date_posted
    date_posted=card_one.find_all('p')[0].text
    date_posted=date_posted.split()[0]
    #technology
    tech=card_one.find('div','job-tags').text
    tech=tech.split()
    tech=tech[3:]
    data=[url_link,job_title,company_name,location,tech,date_posted,job_description]
    return data
    
    
```

```python
#get_info(card_one)
def getAllData():
    data=[]
    for i in outer_point:
        record=get_info(i)
        data.append(record)
    return data
```
